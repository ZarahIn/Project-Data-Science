1. Import Library

2. Read Dataset

```python
import pandas  as pd
from sklearn import datasets

# Memuat dataset digits dari scikit-learn dan mengonversinya menjadi Dataframe
digits = datasets.load_digits()

# Memisahkan fitur dan target
x = digits.data # Fitur  (input)
y = digits.target # Target (output)

# Mengonversi data fitur dan target menjadi DataFrame
df_x = pd.DataFrame(x, columns = digits.feature_names)
df_y = pd.Series(y, name = 'target')

# Gabungkan Fitur dan target dalam satu DataFrame
df = pd.concat([df_x, df_y], axis = 1)

df.head(10)
```

3. Split Data

```python
from sklearn.model_selection import train_test_split

# Membagi data menjadi train dan test
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size = 0.2, random_state = 42)
```

4. Train the Model

```python
from sklearn.ensemble import RandomForestClassifier

# Membuat dan Melatih Random Forest Classifier
model = RandomForestClassifier(n_estimators = 100, random_state = 42)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Memprediksi dan mengevaluasi
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)

print("classification report")
print(classification_report(y_test, y_pred))
print(f"accuracy: {accuracy * 100:.2f}%")
```

5. Data Visualization

a. Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits # Import load_digits here

# Load dataset
digits = load_digits()

# Visualisasi distribusi kelas
sns.countplot(x=digits.target)
plt.title('Distribution of Digits (Target Classes)')
plt.xlabel('Digit')
plt.ylabel('Frequency')
plt.show()
```

b. Feature Correlation Heatmap

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
digits = load_digits()

# Convert fitur menjadi DataFrame
digits_df = pd.DataFrame(digits.data)

# Hitung korelasi antar fitur
corr = digits_df.corr()

# Visualisasi korelasi dengan heatmap
sns.heatmap(corr, annot=False, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
```

c. Sample Image

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load dataset
digits = load_digits()

# Visualisasi beberapa gambar
fig, axes = plt.subplots(2, 5, figsize=(10, 5))  # 2 baris, 5 kolom
for i, ax in enumerate(axes.ravel()):  # Ravel untuk flatten array
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f"Label: {digits.target[i]}")
    ax.axis('off')
plt.suptitle("Sample Images from Digits Dataset")
plt.show()
```
